## Select Columns and Select Rows by Criteria

```python
mtcars.head()
```

```python
#select one column
mtcars['mpg']
```

```python
#select multiple columns
mtcars[['hp', 'mpg']]
```

```python
#subsetting rows based on a condition
mtcars[mtcars['hp'] > 100].head(2)
```

```python
#multiple conditions
mtcars[(mtcars['hp'] > 100) & (mtcars['mpg'] > 20)]
```

```python
#mtcars.loc[rows, columns]
#.loc means locate by name
mtcars.loc[['Mazda RX4'], ['hp']]
```

```python
#.iloc means locate by index aka position (number)
mtcars.iloc[0:5, 0:2]#.head(2)
```

### Some Pandas Functions

- `.head()` and `.tail()` -- show top and bottom rows
- `.describe()`
- `.mean()`, `.std()`, `.min()`, `.max()`
- `.sort_values()`
- `.nlargest()` or `.nsmallest()`

```python
#describe is default numeric summary
mtcars.describe()
```

```python
mtcars.mean()
```

```python
mtcars.min()
```

```python
netflix.head(2)
```

```python
netflix.describe(include = 'all')
```

```python
mtcars.head(2)
```

```python
mtcars.sort_values(by = 'hp', ascending = False).head(2)
```

```python
mtcars.nlargest(2, 'hp')
```

```python
mtcars.nsmallest(2, 'hp')
```

### Categorical vs. Continuous

```python
stroke_df.head()
```

```python
stroke_df.info()
```

```python
stroke_df['gender'].value_counts().plot(kind = 'barh')
```

```python
import seaborn as sns
```

```python
sns.countplot(x='gender', data = stroke_df)
```

```python
sns.countplot(x = 'ever_married', data = stroke_df)
```

```python
sns.displot(stroke_df, x = 'age')
```

```python
sns.displot(stroke_df, x = 'age', kind = 'kde')
```

```python
plt.boxplot(stroke_df['age'])
```

```python
sns.boxplot(x =stroke_df['age'])
```

```python
sns.violinplot(x = stroke_df['age'])
```

```python
fig, ax = plt.subplots(2, 1, figsize = (15, 6))
sns.boxplot(x = stroke_df['age'], ax = ax[0])
sns.kdeplot(data = stroke_df, x = 'age', ax = ax[1])
```

```python
sns.boxplot(x = 'stroke', y = 'age', data = stroke_df)
```

```python
sns.boxplot(x = 'work_type', y = 'age', hue = 'stroke', data = stroke_df)
```

```python
sns.kdeplot(x = 'age', hue = 'stroke', data = stroke_df)
```

```python
sns.countplot(stroke_df['stroke'])
```
